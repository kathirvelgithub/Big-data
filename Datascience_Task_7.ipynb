1.Write a Pandas program to check for inequality of two given DataFrames.

```python
import pandas as pd
import numpy as np

# Create the original DataFrames
data1 = {'W': [68.0, 75.0, 86.0, 80.0, np.nan],
         'X': [78.0, 85.0, np.nan, 80.0, 86.0],
         'Y': [84, 94, 89, 83, 86],
         'Z': [86, 97, 96, 72, 83]}

data2 = {'W': [78.0, 75.0, 86.0, 80.0, np.nan],
         'X': [78, 85, 96, 80, 76],
         'Y': [84, 84, 89, 83, 86],
         'Z': [86, 97, 96, 72, 83]}

df1 = pd.DataFrame(data1)
df2 = pd.DataFrame(data2)

# Display the original DataFrames
print("Original DataFrames:")
print(df1)
print(df2)

# Check for inequality
inequality_result = df1 != df2

# Display the result
print("\nCheck for inequality of the said dataframes:")
print(inequality_result)
```

2.Use the diabetes data set from UCI data set for performing the following:

Apply Univariate analysis:  Frequency,
 Mean,  Median,  Mode,  Variance , Standard Deviation
Skewness and Kurtosis

```python
from google.colab import drive

# Mount Google Drive
drive.mount('/content/drive')
```

```python
import pandas as pd
from scipy.stats import skew, kurtosis

# Load the dataset
url = "/content/drive/MyDrive/diabetes.csv"  # replace with your file path
df = pd.read_csv(url)

# Univariate Analysis: Frequency
frequency = df['BMI'].value_counts()
print("Frequency:\n", frequency)

# Mean, Median, Mode
mean = df.mean()
median = df.median()
mode = df.mode().iloc[0]  # using iloc[0] to get the first mode if there are multiple
print("\nMean:\n", mean)
print("\nMedian:\n", median)
print("\nMode:\n", mode)

# Variance, Standard Deviation
variance = df.var()
std_dev = df.std()
print("\nVariance:\n", variance)
print("\nStandard Deviation:\n", std_dev)

# Skewness and Kurtosis
skewness = df.skew()
kurt = df.kurtosis()
print("\nSkewness:\n", skewness)
print("\nKurtosis:\n", kurt)
```

3.a. Write a NumPy program to convert an array to a float type

  b. Write a NumPy program to create an empty and a full array

  c. Write a NumPy program to convert a list and tuple into arrays

  d. Write a NumPy program to find the real and imaginary parts of an array of complex numbers

```python
# a. Convert an array to a float type:

import numpy as np

# Create an integer array
int_array = np.array([1, 2, 3, 4, 5])

# Convert the array to float
float_array = int_array.astype(float)

print("Original Array (int):", int_array)
print("Converted Array (float):", float_array)
```

```python
#  b. Create an empty and a full array:

import numpy as np

# Create an empty array
empty_array = np.empty((3, 3))
print("Empty Array:")
print(empty_array)

# Create a full array with a specific value
full_array = np.full((2, 4), 7)
print("\nFull Array (with value 7):")
print(full_array)
```

```python
# c. Convert a list and tuple into arrays:

import numpy as np

# Convert a list to an array
list_data = [1, 2, 3, 4, 5]
array_from_list = np.array(list_data)

# Convert a tuple to an array
tuple_data = (6, 7, 8, 9, 10)
array_from_tuple = np.array(tuple_data)

print("Array from List:", array_from_list)
print("Array from Tuple:", array_from_tuple)
```

```python
# d. Find the real and imaginary parts of an array of complex numbers:

import numpy as np

# Create an array of complex numbers
complex_array = np.array([1 + 2j, 3 - 4j, 5 + 6j])

# Find the real and imaginary parts
real_parts = np.real(complex_array)
imaginary_parts = np.imag(complex_array)

print("Complex Array:", complex_array)
print("Real Parts:", real_parts)
print("Imaginary Parts:", imaginary_parts)
```

Task 1,Question 13

4. Apply and explore various plotting functions on Pima Indians Diabetes data
  set for performing the following:

a) Correlation and scatter plots

b) Histograms

c) Three-dimensional plotting

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from mpl_toolkits.mplot3d import Axes3D

# Load the dataset
d= "/content/drive/MyDrive/pima diabetes.csv"  # replace with your file path
df = pd.read_csv(d)

# a) Correlation and scatter plots
correlation_matrix = df.corr()
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Matrix')
plt.show()

# Scatter plots for selected columns
sns.pairplot(df[['Pregnancies', 'Glucose', 'Insulin', 'BMI', 'DiabetesPedigreeFunction', 'Outcome']], hue='Outcome')
plt.suptitle('Scatter Plots')
plt.show()

# b) Histograms
df.hist(figsize=(12, 10))
plt.suptitle('Histograms')
plt.show()

# c) Three-dimensional plotting
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(df['Glucose'], df['BMI'], df['Age'], c=df['Outcome'], cmap='viridis', s=50)
ax.set_xlabel('Glucose')
ax.set_ylabel('BMI')
ax.set_zlabel('Age')
ax.set_title('Three-dimensional Scatter Plot')
plt.show()
```

```python
# c. Create a 3x3 matrix with values ranging from 2 to 10:

import numpy as np

# Create a 3x3 matrix with values ranging from 2 to 10
matrix_3x3 = np.arange(2, 11).reshape(3, 3)

print("3x3 Matrix with Values Ranging from 2 to 10:")
print(matrix_3x3)
```

6.   11. Use the diabetes data set from UCI, Perform Univariate analysis

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
url = "/content/drive/MyDrive/diabetes.csv"  # replace with your file path
df = pd.read_csv(url)

# Univariate analysis: Frequency distribution for the 'Outcome' variable
outcome_counts = df['BMI'].value_counts()
print("Outcome Distribution:")
print(outcome_counts)

# Univariate analysis: Histograms for numeric variables
df.hist(figsize=(12, 10))
plt.suptitle('Histograms of Numeric Variables')
plt.show()

# Univariate analysis: Box plots for numeric variables
df.boxplot(figsize=(12, 8))
plt.title('Box Plots of Numeric Variables')
plt.show()

# Univariate analysis: Summary statistics for numeric variables
summary_statistics = df.describe()
print("Summary Statistics:")
print(summary_statistics)
```
