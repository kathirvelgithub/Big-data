Nested functions

```python
def square():

    def add():

        x = 2
        y = 3
        z = x + y
        return z
    return add()**2
print(square())
```

DEFAULT and FLEXIBLE ARGUMENTS

```python
def f(a, b = 1, c = 2):
    y = a + b + c
    return y
print(f(5))
print(f(5,4,3))
```

```python
def f(*args):
    for i in args:
        print(i)
f(1)
print("")
f(1,2,3,4)
def f(**kwargs):
    """ print key and value of dictionary"""
    for key, value in kwargs.items():
        print(key, " ", value)
f(country = 'spain', capital = 'madrid', population = 123456)
```

zip example

```python
list1 = [1,2,3,4]
list2 = [5,6,7,8]
z = zip(list1,list2)
print(z)
z_list = list(z)
print(z_list)
```

```python
un_zip = zip(*z_list)
un_list1,un_list2 = list(un_zip) # unzip returns tuple
print(un_list1)
print(un_list2)
print(type(un_list2))
```

list comprehension

```python
num1 = [1,2,3]
num2 = [i + 1 for i in num1 ]
print(num2)
```

```python
num1 = [5,10,15]
num2 = [i**2 if i == 10 else i-5 if i < 7 else i+5 for i in num1]
print(num2)
```
