```python
import pandas as pd
n={'a':[1,5,3],'b':[4,8,6]}
nd=pd.DataFrame(n,columns=['a','b'])  #create an dataframe
print(nd,"\n")
nd.sum()
```

```python
nd.cumsum()  #cumsum of dataframe
```

```python
print(nd.min())  #minimum value of the dataframe
print(nd.max())  #maximum value of the dataframe
```

```python
nd.describe()   #describe of the dataframe
```

```python
nd.median()  #median of dataframe
```

```python
f=lambda x:x**2 # create lambda function to square
nd.apply(f)
```

```python
h=lambda y:y*2  # create lambda function to multiple by 2
nd.apply(h)
```

```python
t=lambda x:x+15  # create lambda function to addition by 15
nd.applymap(t)
```

```python
nd.sort_index()
```

```python
nd.sort_values(by='a')
```
